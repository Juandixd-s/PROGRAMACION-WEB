## Ejercicio

Escribir un programa que permita gestionar la base de datos de clientes de una empresa. Los clientes se guardarán en un diccionario en el que la clave de cada cliente será su NIF, y el valor será otro diccionario con los datos del cliente (nombre, dirección, teléfono, correo, preferente), donde preferente tendrá el valor `True` si se trata de un cliente preferente. El programa debe preguntar al usuario por una opción del siguiente menú: (1) Añadir cliente, (2) Eliminar cliente, (3) Mostrar cliente, (4) Listar todos los clientes, (5) Listar clientes preferentes, (6) Terminar. En función de la opción elegida el programa tendrá que hacer lo siguiente:

1. Preguntar los datos del cliente, crear un diccionario con los datos y añadirlo a la base de datos.
2. Preguntar por el NIF del cliente y eliminar sus datos de la base de datos.
3. Preguntar por el NIF del cliente y mostrar sus datos.
4. Mostrar lista de todos los clientes de la base datos con su NIF y nombre.
5. Mostrar la lista de clientes preferentes de la base de datos con su NIF y nombre.
6. Terminar el programa.

## Solución

```python
import json
import os

if os.path.exists("clientes.json"):
    with open("clientes.json", "r", encoding="utf-8") as archivo:
        clientes = json.load(archivo)
else:
    clientes = {}


def guardar_datos():
    with open("clientes.json", "w", encoding="utf-8") as archivo:
        json.dump(clientes, archivo, indent=4, ensure_ascii=False)


while True:
    print("\n====== MENU CLIENTES ======")
    print("1. Añadir cliente")
    print("2. Eliminar cliente")
    print("3. Mostrar cliente")
    print("4. Listar todos los clientes")
    print("5. Listar clientes preferentes")
    print("6. Terminar")

    opcion = input("Seleccione una opción: ")

    # ----------- AÑADIR CLIENTE -----------
    if opcion == "1":
        nif = input("Ingrese NIF: ")
        nombre = input("Ingrese nombre: ")
        direccion = input("Ingrese dirección: ")
        telefono = input("Ingrese teléfono: ")
        correo = input("Ingrese correo: ")
        preferente = input("¿Es cliente preferente? (True/False): ")

        lista_compras = []
        cantidad = int(input("¿Cuántas compras desea registrar?: "))

        for i in range(cantidad):
            compra = input("Ingrese producto comprado: ")
            lista_compras.append(compra)

        clientes[nif] = {
            "nombre": nombre,
            "direccion": direccion,
            "telefono": telefono,
            "correo": correo,
            "preferente": preferente == "True",
            "lista_compras": lista_compras
        }

        guardar_datos()
        print("Cliente agregado correctamente.")

    # ----------- ELIMINAR CLIENTE -----------
    elif opcion == "2":
        nif = input("Ingrese NIF del cliente a eliminar: ")

        if nif in clientes:
            del clientes[nif]
            guardar_datos()
            print("Cliente eliminado correctamente.")
        else:
            print("Cliente no encontrado.")

    # ----------- MOSTRAR CLIENTE -----------
    elif opcion == "3":
        nif = input("Ingrese NIF del cliente a mostrar: ")

        if nif in clientes:
            print("\nDatos del cliente:")
            print("NIF:", nif)
            for clave, valor in clientes[nif].items():
                print(clave, ":", valor)
        else:
            print("Cliente no encontrado.")

    # ----------- LISTAR TODOS -----------
    elif opcion == "4":
        print("\nLISTA GENERAL DE CLIENTES")
        for nif, datos in clientes.items():
            print(nif, "-", datos["nombre"])

    # ----------- LISTAR PREFERENTES -----------
    elif opcion == "5":
        print("\nCLIENTES PREFERENTES")
        for nif, datos in clientes.items():
            if datos["preferente"] == True:
                print(nif, "-", datos["nombre"])

    # ----------- TERMINAR -----------
    elif opcion == "6":
        print("Programa finalizado.")
        break

    else:
        print("Opción inválida.")
```
