## Ejercicio

Escribir un programa que pregunte una fecha en formato `dd/mm/aaaa` y muestre por pantalla la misma fecha en formato `dd de <mes> de aaaa` donde `<mes>` es el nombre del mes.

```python
meses = {1:'enero', 2:'febrero', 3:'marzo', 4:'abril', 5:'mayo', 6:'junio', 7:'julio', 8:'agosto', 9:'septiembre', 10:'octubre', 11:'noviembre', 12:'diciembre'}
fecha = input('Introduce una fecha en formato dd/mm/aaaa: ')
fecha = fecha.split('/')
print(fecha[0], 'de', meses[int(fecha[1])], 'de', fecha[2])
fecha_de_cumpleaños = input("introduce tu fecha de cumpleaños dd/mm/aaaa:")
fecha_de_cumpleaños = fecha_de_cumpleaños.split('/')
print(fecha_de_cumpleaños[0], 'de', meses[int(fecha_de_cumpleaños[1])], 'de', fecha_de_cumpleaños[2])
fecha_de_reunion = input("introduce tu fecha de reunión dd/mm/aaaa:")
fecha_de_reunion = fecha_de_reunion.split('/')
print(fecha_de_reunion[0], 'de', meses[int(fecha_de_reunion[1])], 'de', fecha_de_reunion[2])
```
